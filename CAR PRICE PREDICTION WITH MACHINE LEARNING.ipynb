```python
# importing libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv("D:\Downloads\Car price.csv")
```

```python
print(Data.isnull().sum())
```

```python
#Here we can see there are no null values.
#Now we will delete the columns which we don't neeed
```

```python
#Now we will create feature(independent) and target(dependant variable) variable
```

```python
#next we will be spliting the dataset into train and test sets
```

```python
categorical_cols = ['CarName', 'fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem']
ct = ColumnTransformer([('encoder', OneHotEncoder(), categorical_cols)], remainder='passthrough')
X = ct.fit_transform(Data)
```

```python
y = Data['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#lets check out the linear regression model for the dataset
```

```python
model = LinearRegression()
```

```python
#fitting the train data 
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

```python
plt.scatter(y_test, y_pred, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Linear Regression')
```

```python
#By this plot we can say that dependent and independent variables are highly correleated
```

```python
# Evaluate the performance of the model using the R-squared value
```

```python
r2_score = r2_score(y_test, y_pred)
print('R-squared score:', r2_score)
```
